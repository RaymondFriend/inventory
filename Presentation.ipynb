# Total ECoS Footprint

Eberly produced...

## <center> 24,111 MtCO$_2$e </center>

... of greenhouse gas emissions during 2019

That's equivalent to:

- Driving $\mathbf{5,250}$ **passenger vehicles** continuously for a year

- Burning $\mathbf{26,700,000}$ **lbs of coal**

- Powering $\mathbf{2,900}$ **U.S. homes** for a year

It's also equivalent to:

- The carbon sequestered in $\mathbf{29,500}$ **acres of forest** over a year

- The amount avoided by running $\mathbf{5}$ **windmills** for a year

- Powering just $\mathbf{0.006}$ **coal-fired power plants** for a year
